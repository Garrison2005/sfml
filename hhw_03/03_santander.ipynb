```python
import pandas as pd
```

```python
train = pd.read_csv('03_santander_train.csv', index_col='ID')
```

```python
train.head()
```

```python
test = pd.read_csv('03_santander_test.csv', index_col='ID')
```

```python
test.head()
```

```python
y = list(train['TARGET'])
```

```python
train = train.drop('TARGET', axis=1)
```

```python
train.info()
```

```python
train.info(verbose=True, memory_usage=True, null_counts=True) # Если данных очень много
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train, y, test_size=0.33, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
%matplotlib inline
from matplotlib import pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
dtc = DecisionTreeClassifier(max_depth=5)
dtc.fit(X_train, y_train)
```

```python
# Подберем лучший max_depth, исходя из максимизации ROC_AUC_score
N = 15 # Максимальное кол-во делений дерева
s = 0
md = 0
i = 1
while i<=N:
    dtc = DecisionTreeClassifier(max_depth=i)
    dtc.fit(X_train, y_train)
    predict = dtc.predict_proba(X_test)
    if roc_auc_score(y_test, predict[:,1])>s:
        s = roc_auc_score(y_test, predict[:,1])
        md = i
    i+=1
print ('ROC_AUC_score: ',s,'  Лучший max_depth: ',md)
```

```python
i = 7
dtc = DecisionTreeClassifier(max_depth=i)
dtc.fit(X_train, y_train)
predict = dtc.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(y_test, predict[:,1])
plt.figure(figsize=(5, 5))
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
# Подберем лучшее n_neighbors, исходя из максимизации ROC_AUC_score
N = 100 # Максимальное кол-во cоседних точек
s = 0
md = 0
i = 1
while i<=N:
    knc = KNeighborsClassifier(n_neighbors=i,n_jobs= 12)
    knc.fit(X_train, y_train)
    predict = knc.predict_proba(X_test)
    if roc_auc_score(y_test, predict[:,1])>s:
        s = roc_auc_score(y_test, predict[:,1])
        md = i
    i+=1
print ('ROC_AUC_score: ',s,'  Лучшее n_neighbors: ',md)
```

```python
i = 63 # при N=50: 49; при N=70: 63, при Т=100: 99.
knc = KNeighborsClassifier(n_neighbors=i,n_jobs= 12)
knc.fit(X_train, y_train)
predict = knc.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(y_test, predict[:,1])
plt.figure(figsize=(5, 5))
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```

```python
# Подберем лучшее C, исходя из максимизации ROC_AUC_score
N = 5 # Максимальное значение С
s = 0
md = 0
i = 0.1
while i<=N:
    lr = LogisticRegression(C = i, penalty='l1') 
    lr.fit(X_train, y_train)
    predict = lr.predict_proba(X_test)
    if roc_auc_score(y_test, predict[:,1])>s:
        s = roc_auc_score(y_test, predict[:,1])
        md = i
    i+=0.1
print ('penalty = l1')
print ('ROC_AUC_score: ',s,'  Лучшее n_neighbors: ',md)
```

```python
N = 5 
s = 0
md = 0
i = 0.1
while i<=N:
    lr = LogisticRegression(C = i,penalty='l2') 
    lr.fit(X_train, y_train)
    predict = lr.predict_proba(X_test)
    if roc_auc_score(y_test, predict[:,1])>s:
        s = roc_auc_score(y_test, predict[:,1])
        md = i
    i+=0.1
print ('penalty = l2')
print ('ROC_AUC_score: ',s,'  Лучшее n_neighbors: ',md)
```

```python
# Выбираем l1
i = 3.6
lr = LogisticRegression(C = i,penalty='l1') 
lr.fit(X_train, y_train)
predict = lr.predict_proba(X_test)
fpr, tpr, thresholds = roc_curve(y_test, predict[:,1])
plt.figure(figsize=(5, 5))
plt.plot(fpr, tpr)
plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
```
